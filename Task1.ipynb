```python
import pandas as pd
```

```python
df = pd.read_csv('/content/netflix_titles.csv')
```

```python

df['cast'].fillna('Unavailable', inplace=True)
df['country'].fillna('Unavailable', inplace=True)
```

```python
df.info()
```

```python
# If all 'show_id' values are numeric strings
df['show_id'] = pd.to_numeric(df['show_id'], errors='coerce')
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
# Convert date formats
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce', dayfirst=True)
```

```python
# Rename columns
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python

# Fix data types
df['release_year'] = df['release_year'].astype(int)
```
